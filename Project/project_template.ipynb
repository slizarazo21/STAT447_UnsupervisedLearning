```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, silhouette_samples, silhouette_score
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn_extra.cluster import KMedoids
from sklearn.manifold import TSNE
from gower import gower_matrix
```

This dataset has a correlation in gene expression between specific genes (the ones in the columns) and all the genes that are available in a data set, about 20000 genes. I filtered out those genes with a median correlation value across all the genes specific genes between -0.15 and 0.15. Leaving only genes with high and low correlation. 

This data set is just the expression levels of a lot of genes across patients with Kidney Cancer. I think that in this dataframe are patients with and without the disease. We can explore that. We can also filter the data frame further and account for different types of genes, let's say genes that are linked to cancer, linked to immune responses and so on to further explore the identity of these genes. 

```python
data4 = pd.read_csv('test.csv')
data4.columns
```

```python
data = data4.copy()
data = data.sample(5000, random_state = 447)

### selecting columns we care for

data = data.iloc[:,3:25]
data.dtypes
```

```python
categorical = [data.columns[0]] + list(data.columns[2:19]) + [data.columns[21]]

for i in categorical:
    data[i] = data[i].astype('object')
    
data.dtypes

data = data.dropna()
```

```python
dist_mat = gower_matrix(data)
```

```python
for perp in [5,10, 20, 30, 40, 50]:
    for rs in [14, 22]:
        tsne = TSNE(n_components=2, perplexity=perp, random_state=rs, metric='precomputed',init='random')
        data_tsne = tsne.fit_transform(dist_mat)
        df_tsne = pd.DataFrame(data_tsne, columns=['x_projected', 'y_projected'])
        df_combo = pd.concat([data, df_tsne], axis=1)
        sns.scatterplot(x='x_projected',y='y_projected', data=df_combo)
        plt.title('t-SNE Plot with Perplexity Value %s and Random State %s' %(perp, rs))
        plt.show()
    print('--------------------------------------------')
```
