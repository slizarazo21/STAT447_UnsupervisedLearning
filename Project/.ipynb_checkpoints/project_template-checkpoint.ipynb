```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, silhouette_samples, silhouette_score
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn_extra.cluster import KMedoids
from sklearn.manifold import TSNE
```

This dataset has a correlation in gene expression between specific genes (the ones in the columns) and all the genes that are available in a data set, about 20000 genes. I filtered out those genes with a median correlation value across all the genes specific genes between -0.15 and 0.15. Leaving only genes with high and low correlation. 

```python
data1 = pd.read_csv('BRCA.Breast_invasive_carcinoma.rnaseqv2__illuminahiseq_rnaseqv2__unc_edu__Level_3__RSEM_genes_normalized__data.data.txt_corr.txt', sep= ',')
data1.head()

select_col = [col for col in data1.columns if '_pv' not in col]

data11 = data1[select_col]
data11 = data11[data11.columns[2:-1]]
data11 = data11.dropna()

median = data11.median(axis = 1)

data11 = data11[(median <= -0.15) | (median >= 0.15)]

data11.shape


for i in [5,10,20,30,40,50]:
    for k in [55, 120]:
        tsne = TSNE(n_components = 2, perplexity = i, random_state = k)
        datsne = tsne.fit_transform(data11)
        df_t = pd.DataFrame(datsne, columns = ['x_p', 'y_p'])
        df_c = pd.concat([data11.T, df_t], axis = 1)
        sns.scatterplot(x = 'x_p', y = 'y_p', data = df_c)
        plt.title('t-SNE, Perplexity (%s), Random State (%s)' %(i,k))
        plt.show()
    print('-\\-\\-\\')
```
