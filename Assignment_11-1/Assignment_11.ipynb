## 3. Gaussian Mixture Models Parameter Selection

We would like to use Gaussian Mixture models to cluster this dataset. However, we are currently unsure as to how many clusters we should ask for.

### 3.1. Parsimonious Gaussian Mixture Model

We would first like to assess the number of clusters $k$ which will yield a parsimonious Gaussian mixture model for this dataset. 

#### 3.1.1. Reliability of AIC for Selecting a Parsimonious Model

For what values of the cluster number $k$ can we rely on the AIC score to produce useful insights regarding the parsimonious model? In general, can we rely on the AIC score at all with this dataset?



Let's assume that we want to define a total of 10 cluster, one for each letter. Then the total number of parameters that our model needs to estimate will be equal to k(1 + d + d^2). Where d = 784 (total number of attributes) The data frame has a total of 500 observations. BY RULE OF THUMB we can use AIC when n/np > 40. In this case we have 500/15710  which is less than 40. Therefore using AIC might not be the best idea in this case

#### 3.1.2. AIC and BIC Score Plots

Create two plots.
1. One plot should plot the AIC score of a Gaussian mixture model clustering with k=1, k=2, ..., k=19 clusters respectively.
2. The other plot should plot the BIC score of a Gaussian mixture model clustering with k=1, k=2, ..., k=19 clusters respectively.

Use a random state of 101 for all of these clusterings.

````python
aics = []
bics = []

for k in range(1,20):
    #Fit the GMM model
    gm = GaussianMixture(n_components= k, random_state = 101)
    gm.fit(df_num)
    ## AIC and BIC
    aics.append(gm.aic(df_num))
    bics.append(gm.bic(df_num))
    
## plot
fig, ax = plt.subplots(1, 2, figsize = (30,6))

# AIC
ax[0].plot(range(1,20), aics, color = 'red')
ax[0].set_xticks(range(1, 20))
ax[0].set_xlabel('k', fontsize=16)
ax[0].set_title('AIC', fontsize=16)

#BIC
ax[1].plot(range(1, 20), bics, color = 'blue')
ax[1].set_xticks(range(1, 20))
ax[1].set_xlabel('k', fontsize=16)
ax[1].set_title('BIC', fontsize=16)
````

#### 3.1.3 Interpretation

Use the AIC score to select the number of clusters that will yield a parsimonious Gaussian mixture model. Use the BIC score to select the number of clusters that will yield a parsimonious Gaussian mixture model.

Based on the plots, the optimal number of cluster for AIC is 2 and for BIC is 1.

#### 3.1.4. Usefulness

Given that our goal is to determine the underlying clusters of this dataset, particularly those that the t-SNE plot is suggesting might exist, do we think that the cluster number $k$ suggested by the AIC or BIC score are useful?

No, the scores are not useful. The goal was to at least be able to see 10 clusters one for each letter but AIC or BIC did not return any of this optimal number of clusters. 

## 4. Gaussian Mixture Model with k=10 clusters

### 4.1. Cluster the Data
Cluster the dataset into k=10 clusters using Gaussian mixture models. Use a random state of 101.

````python
gm = GaussianMixture(n_components= 10, random_state=101)

gm.fit(df_num)
````

### 4.2. t-SNE Plots

Then, plot k=10 t-SNE plots. In the kth plot, you should color-code the objects by their kth cluster membership weight (ie. their kth cluster posterior value).

````python
W = gm.predict_proba(df_num)
cluster_columns=['Cluster %s Membership Score'%(str(i)) for i in range(0,10)]
df_membership=pd.DataFrame(W, columns=cluster_columns)
df_membership.round(4)
````

````python
fig, ax = plt.subplots(2, 5, figsize=(20, 10))
plt.suptitle('Clustering: %s clusters'%(10), fontsize=14)
for i in range(0,10):

    #Create a column with cluster i membership scores
    df_combo['Cluster %s Membership Score'%str(i)]=df_membership['Cluster %s Membership Score'%(str(i))]

    #Plot the points, color coded by the membership score.
    points = plt.scatter(df_combo["x_projected"], df_combo["y_projected"],
                 c=df_combo['Cluster %s Membership Score'%str(i)],
                         s=30, cmap="viridis",
                        vmin=0,
                        vmax=1)
    ax[int(np.floor(i/5)),int(i%5)].scatter(df_combo["x_projected"], df_combo["y_projected"],
                 c=df_combo['Cluster %s Membership Score'%str(i)],
                                            s=30, cmap="viridis",
                                           vmin=0,
                                           vmax=1)

    fig.colorbar(points, ax=ax[int(np.floor(i/5)),int(i%5)])
    ax[int(np.floor(i/5)),int(i%5)].set_title('Cluster %s Membership Scores'%(str(i)))
plt.show()
````

### 4.3. Visualize the Cluster Means

Finally, visualize the 10 cluster mean images below.

````python
print('Means:')
print()

#Let's visualize and describe the cluster centroids for this clustering
#Set up the plot
fig, ax = plt.subplots(1, 10, figsize=(12, 2))
plt.suptitle('Images of the Cluster/Gaussian Distribution Means')

for j in range(0,10):
    #Turn the centroids into a dataframe
    df_centroids=pd.DataFrame(gm.means_, columns=df_num.columns)
    #First, reshape the flattened 784 pixel array into an 28-by-28 numpy array and unscale the data (ie. multiply by 255)
    object_image = np.array(df_centroids.iloc[j]).reshape((28, 28))*255

    #Then show the image
    ax[j].imshow(object_image, cmap='cubehelix')
    ax[j].grid(False)
    ax[j].set_xticks([])
    ax[j].set_yticks([])
    ax[j].set_title('Mean %s'%str(j))
plt.show()
````

````python
### Expectation step
### Calculating the pdf for each observation times the prior for each cluster
mvn0 = multivariate_normal(mean=m0, cov=s0)
mvn1 = multivariate_normal(mean=m1, cov=s1)

# Apply the pdf function to each row for the first Gaussian
data['clust 0'] = data.apply(lambda row: mvn0.pdf([row['x1'], row['x2']]), axis=1)

# Similarly for the second Gaussian
data['clust 1'] = data.apply(lambda row: mvn1.pdf([row['x1'], row['x2']]), axis=1)

## Multiplying for the prior for each Gaussian
data['clust 0'] = data['clust 0']*0.4
data['clust 1'] = data['clust 1']*0.6

### Calculating posteriors for each cluster
data['clust 0 post'] = data['clust 0'] / (data['clust 0'] + data['clust 1'])
data['clust 1 post'] = data['clust 1'] / (data['clust 0'] + data['clust 1'])

data
````
