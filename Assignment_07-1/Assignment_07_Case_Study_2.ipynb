## 4. Aggomerative Hierarchical Clustering

### 4.1. Faithful Preservation of the Original Pairwise Distances

Suppose we would like to cluster this datset using an agglomerative hierarchical clustering algorithm. Which of the four linkage functions (single, complete, average, or Ward's) will produce a dendrogram that most faithfully preserves the pairwises distances of the points in the orginal dataset?

`

```python
ccc_list=[]
linkage_list=['single', 'complete', 'ward', 'average']

for link in linkage_list:
    #Create the dendrogram
    dm = pdist(df_num1, metric='euclidean')
    Z = linkage(dm, method=link)
    
    #Calculate the cophenetic correlation coefficient of the linkage function dendrogram
    ccc, _ = cophenet(Z, dm)
    ccc_list.append(ccc)

#Plot the results
plt.plot(linkage_list, ccc_list)
plt.title('Cophenetic Correlation Coefficient')
plt.xlabel('Linkage Function')
plt.ylabel('Cophenetic Correlation Coefficient')
plt.show()
```

```python
for i in ['single', 'complete', 'ward', 'average']:
    mat = pdist(df_num1, metric = 'euclidean')
    link = linkage(mat, method = i)
    
    print(i)
    fig, ax = plt.subplots(figsize = (25,30))
    d = dendrogram(link, orientation = 'right', labels = df['Wine Class'].array, ax = ax)
    ax.set_xlabel('Dissimilarity')
    ax.set_ylabel('States')
    plt.yticks(fontsize = 15)
    plt.show()
```

Average linkage as it has the highest cophenetic correlation, even though when we look at the different hierarchical clusters, we observe that the clusters are better observed with the ward linkage

### 5.1. Clustering with Most t-SNE Corroboration

Next, find the clustering in your dendrogram from 4.2 with the lowest amount of clusters that has the *most* amount of association with the $k$ "main clusters" that you identified in your t-SNE plots.

### 5.6. Cluster Sorted Similarity Matrix Correlation

#### 5.6.1. Correlation
Scale this clusterted sorted similarity matrix (using the method we discussed in class) and calculate the correlation of this scaled matrix to the "ideal" scaled cluster sorted similarity matrix which has the same number of clusters and the same number of objects within each of these clusters.

```python
# Minimum
min_dist=np.min(dist_mat1)

# Maximum
max_dist=np.max(dist_mat1)

#Similarity matrix
sim_mat=1-(dist_mat1-min_dist)/(max_dist-min_dist)
sim_mat
```

```python
from scipy.linalg import block_diag

#Count the number of objects in each cluster, and sort by cluster label.
cluster_sizes=df_combo['predicted_cluster'].value_counts().sort_index()
print('Cluster Sizes')
print(cluster_sizes)
print()

#Create a list of square matrices with all ones, each matrix is the size of a cluster.
block_list=[]
for i in cluster_sizes:
    print('Cluster Size:',i)
    block=np.ones((i,i)).tolist()
    block_list.append(block)

#Put these square matrices on the block diagonal of an otherwise 0-matrix
print()
ideal_sim_mat=block_diag(block_list[0], block_list[1], block_list[2], block_list[3],
                        block_list[4], block_list[5], block_list[6], block_list[7])
ideal_sim_mat
```

```python
flat_sim_mat=sim_mat.flatten()
print('Flattened Actual Scaled Similarity Matrix')
print(flat_sim_mat)
len(flat_sim_mat)    
flat_ideal_sim_mat=ideal_sim_mat.flatten()
print('Flattened Ideal Scaled Similarity Matrix')
print(flat_ideal_sim_mat)
```

```python
np.corrcoef(flat_sim_mat,flat_ideal_sim_mat)
```
